- 常用算法
    - 穷举法 -又称为暴力破解法，对所有可能性就行验证，直到找到正确答案
    - 贪婪法 - 在对问题求解是，总是做出当前看来最好的选择，不追求最优解，快速找到满意答案
    - 分治法 - 把一个复杂的问题分成两个或更多的相同或相似的子问题，再把子问题分成更小的子问题，知道可以直接求解的程度，最后将子问题的解进行合并得到原问题的解
    - 回溯法 - 回溯法又称为试探法，按选优条件先前搜索，当搜索到某一步发现原先这样选择不能达到目标时，就退回到上一步重新选择
    - 动态规划法 - 基本思想也是将待求解问题分解成若干个子问题，先求解并保存子问题的解，避免大量重复运算

## 穷举法
- 例：百鸡百钱<br>
    > 公鸡5元一只，母鸡3元一只，小鸡一元3只
    用100元买一百只鸡，问公鸡/母鸡/小鸡各多少只
- 例：五人分🐟<br>
    >A、B、C、D、E五人在某个夜晚合伙去捕鱼，最后疲惫不堪各自去睡觉
    第二天A第一个醒来，他将鱼分为五份，把多余的一条扔掉，拿走自己的一份
    B第二个醒来，也将鱼分成了5份，把多余的一条扔掉，拿走自己的一份
    然后C、D、E依次醒来也按同样的方式分鱼
    问：他们至少捕了多少鱼

```python
for x in range(20):
    for y in range(34):
        for z in range(100):
            if x + y + z == 100 and x * 5 + y * 3 + z // 3 == 100:
                print(x, y, z)

```

## 贪婪法
- 在对问题求解时，总是做出在当前看来做好的选择，不追求最优解，快速找到满意的答案
- 例：背包问题
>假设小偷有一个背包，最多能装20公斤赃物，他闯入一户人家，发现如下表所示的物品。很显然，他不能把所有物品都装进背包，所以必须确定拿走哪些物品，留下哪些物品。

| 名称 | 价格(\$)| 重量(Kg) |
| ---- | ------ | ------- |
| 电脑 | 200 | 20 |
| 收音机 | 20 | 4 |
| 钟 | 175 | 10 |
| 花瓶 | 50 | 2 |
| 书 | 10 | 1 |
| 油画 | 90 | 9 |

```python
class Thing():
    
    def __init__(self, name, price, weight):
        self.name = name
        self.weight = weight
        self.price = price
    
    @property
    def value(self):
        return self.price / self.weight


def main():
    max_weight = int(input('小偷背包的承重：'))
#     num_of_thing = int(input('房间内的物品数量：'))
    all_things = [
        Thing('电脑', 200, 20),
        Thing('收音机', 20, 4),
        Thing('钟', 175, 40),
        Thing('花瓶', 50, 2),
        Thing('书', 10, 1),
        Thing('油画', 90, 9)
    ]
#     for thing in range(num_of_thing):
#         all_things.append(Thing(*input_thing()))
    all_things.sort(key=lambda x:x.value, reverse=True)
#     print(all_things)
    total_weight = 0
    total_price = 0
    for thing in all_things:
        if total_weight + thing.weight <= max_weight:
            total_weight += thing.weight
            total_price += thing.price
            print(f'小偷偷走了{thing.name},价值{thing.price}')
    print('小偷共偷走了价值%d的物品' % total_price)
    
    
if __name__ == '__main__':
    main()
```

## 分治法
- 快速排序
>选择枢轴对元素进行划分，小的放在左边，大的放在右边

```python
def quick_sort(origin_item, comp=lambda x, y: x <= y):
    items = origin_item[:]
    _quick_sort(items, 0, len(items) - 1, comp)
    return items


def _quick_sort(items, start, end, comp):
    if start < end:
        pos = _partition(items, start, end, comp)
        _quick_sort(items, start, pos - 1, comp)
        _quick_sort(items, pos + 1, end, comp)


def _partition(items, start, end, comp):
    pivot = items[end]
    i = start
    for j in range(start, end):
        if comp(items[j], pivot):
            items[i], items[j] = items[j], items[i]
            i += 1
    items[i], items[end] = items[end], items[i]
    return i


def main():
    origin_item = [56, 34, 67, 235, 45, 1, 62, 6, 123, 45, 254, 125, 2, 8, 34, 6]
    print(origin_item)
    print(quick_sort(origin_item))


if __name__ == '__main__':
    main()
```

## 回溯法
- 又称为试探法，按选优条件向前探索，当探索到某一步，发现原先选择的并非最优或达不到目标，就退回一步重新选择
- 比较经典的问题有骑士巡逻、八皇后和迷宫寻路等。
- 例：骑士巡逻
> 将一个国际象棋的骑士（或称马）放在棋盘上，有什么路径能使它走遍棋盘上每一格呢

```python
SIZE = 5
total = 0


def print_board(board):
    for row in board:
        for col in row:
            print(str(col).center(4), end='')
        print()


def patrol(board, row, col, step=1):
    if 0 <= row < SIZE and \
            0 <= col < SIZE and \
            board[row][col] == 0:
        board[row][col] = step
        if step == SIZE * SIZE:
            global total
            total += 1
            print(f'第{total}种走法: ')
            print_board(board)
        patrol(board, row - 2, col - 1, step + 1)
        patrol(board, row - 1, col - 2, step + 1)
        patrol(board, row + 1, col - 2, step + 1)
        patrol(board, row + 2, col - 1, step + 1)
        patrol(board, row + 2, col + 1, step + 1)
        patrol(board, row + 1, col + 2, step + 1)
        patrol(board, row - 1, col + 2, step + 1)
        patrol(board, row - 2, col + 1, step + 1)
        board[row][col] = 0


def main():
    board = [[0] * SIZE for _ in range(SIZE)]
    patrol(board, SIZE - 1, SIZE - 1)


if __name__ == '__main__':
    main()
```

## 动态规划法
- 适用于重叠子问题和最优子结构性质的问题，使用动态规划法所耗时间远远小于朴素解法（用空间换时间）
- 例：子列表元素之和最大值（使用动态规划可以避免二次循环）
> 说明：子列表是指列表种索引连续的元素构成的列表；
    列表中的元素是int类型，可以包含正整数/0/负整数
    成功内需程序输入列表中的元素，输出子列表元素和的最大值
         输入：1 -2 3 5 -3 2
         输出：8
         输入：0 -2 3 5 -1 2
         输出：9
         输入：-9 -2 -3 -5 -3
         输出：-2

```python
def main():
    items = list(map(int, input().split()))
    size = len(items)
    overall, partial = {}, {}
    overall[size - 1] = partial[size - 1] = items[size - 1]
    for i in range(size - 2, -1, -1):
        partial[i] = max(items[i], partial[i + 1] + items[i])
        overall[i] = max(partial[i], overall[i + 1])
    print(overall[0])


if __name__ == '__main__':
    main()
```
